```python
#導入常用函式庫
import numpy as np 
import pandas as pd
import warnings
warnings.simplefilter(action='ignore')
#1)將資料讀取進來
df = pd.read_csv("train_feature.csv",encoding='big5')
df1 = pd.read_csv('train_label.csv',encoding='big5')
#檢視資料
df.tail(20)
```

```python
#查看欄位的數值型態及是否有缺失值需要進行處理
df.info()
df.isnull().sum()
```

```python
df.drop(['Index','主要建材','主要用途','交易年月日','交易標的','備註','土地區段位置/建物區段門牌','建築完成年月','移轉層次','編號','車位類別','非都市土地使用分區','非都市土地使用編定','lat','lng','總樓層數'],axis=1,inplace=True)
df1.drop(['Index'],axis=1,inplace=True)
df.info()
```

```python
df["交易筆棟數"]=df["交易筆棟數"].str.replace('土地','')
df["交易筆棟數"]=df["交易筆棟數"].str.replace('建物',',')
df["交易筆棟數"]=df["交易筆棟數"].str.replace('車位',',')
df[['土地', '建物', '車位']] = df['交易筆棟數'].str.split(',', expand=True)
df.drop(['交易筆棟數'],axis=1,inplace = True)
df["都市土地使用分區"].fillna('其他',inplace = True)
df["num_of_bus_stations_in_100m"].fillna(0,inplace = True)
df["income_avg"].fillna(df["income_avg"].mean(),inplace = True)
df["income_var"].fillna(df["income_var"].mean(),inplace = True)
df["location_type"].fillna('ROOFTOP',inplace = True)
df["low_use_electricity"]=df["low_use_electricity"].str.replace('%','')
df["low_use_electricity"]=df["low_use_electricity"].astype(float)
df["low_use_electricity"]=df['low_use_electricity'].apply(lambda x: x / 100)
df["low_use_electricity"].fillna(df["low_use_electricity"].mean(),inplace = True)
df["nearest_tarin_station"].fillna('先嗇宮',inplace = True)
df["nearest_tarin_station_distance"].fillna(0,inplace = True)
df.columns
```

```python
df.info()
df.isnull().sum()
```

```python
columns_to_transform = ['建物現況格局-隔間','建物型態', '有無管理組織','都市土地使用分區','鄉鎮市區','location_type','nearest_tarin_station']
df = pd.get_dummies(df, columns=columns_to_transform)
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, df1, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
score = model.score(X_test, y_test)
score
```
